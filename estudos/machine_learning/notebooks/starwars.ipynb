```python
import pandas as pd

df = pd.read_parquet("../data/dados_clones.parquet")
df
```

```python
from sklearn import tree

y = df[["Status "]]
X = df.drop(columns=["Status ", "p2o_master_id"])

X.replace({
    "Tipo 5": 5, "Tipo 4": 4, "Tipo 3": 3, "Tipo 2": 2, "Tipo 1": 1,
    'Yoda': 1, 'Shaak Ti': 2, 'Obi-Wan Kenobi': 3, 'Aayla Secura': 4, 'Mace Windu': 5
}, inplace=True)

X.columns
```

```python
model = tree.DecisionTreeClassifier()
model.fit(X, y)
```

```python
import matplotlib.pyplot as plt
    

plt.figure(dpi=400)
tree.plot_tree(model, feature_names=X.columns, 
class_names=model.classes_, filled=True, max_depth=3)
```

```python
X.drop(columns=["General Jedi encarregado"], inplace=True)
modelWithoutGeneralJedi = tree.DecisionTreeClassifier()
modelWithoutGeneralJedi.fit(X, y)
```

```python
plt.figure(dpi=400)
tree.plot_tree(modelWithoutGeneralJedi, feature_names=X.columns, 
class_names=modelWithoutGeneralJedi.classes_, filled=True, max_depth=3)
```
